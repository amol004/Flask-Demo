```python
%%writefile streamlit_app.py
import streamlit as st
import pickle
import numpy as np

st.set_page_config(page_title="Crop Recommendation App", layout="centered")

# Load the trained model
try:
    # Assuming 'model1.pkl' is the correct model for crop recommendation
    model = pickle.load(open('model1.pkl', 'rb'))
except FileNotFoundError:
    st.error("Error: model1.pkl not found. Make sure the model is saved correctly in the same directory.")
    st.stop()

st.title("ð± Crop Recommendation System ð±")
st.markdown("Enter the environmental parameters to get a crop recommendation.")

# Input fields for features
st.sidebar.header("Input Parameters")

N = st.sidebar.slider("Nitrogen (N)", 0.0, 140.0, 90.0, step=0.1)
P = st.sidebar.slider("Phosphorus (P)", 5.0, 145.0, 42.0, step=0.1)
K = st.sidebar.slider("Potassium (K)", 5.0, 205.0, 43.0, step=0.1)
temperature = st.sidebar.slider("Temperature (Â°C)", 8.0, 43.0, 20.88, step=0.01)
humidity = st.sidebar.slider("Humidity (%)", 14.0, 99.0, 82.00, step=0.01)
ph = st.sidebar.slider("pH Value", 3.5, 9.9, 6.50, step=0.01)
rainfall = st.sidebar.slider("Rainfall (mm)", 20.0, 298.0, 202.93, step=0.01)

# Prepare features for prediction
features = np.array([[N, P, K, temperature, humidity, ph, rainfall]])

if st.sidebar.button("Recommend Crop"): # Moved button to sidebar
    if model is not None:
        prediction = model.predict(features)[0]
        st.success(f"The recommended crop for these conditions is: **{prediction.upper()}**")
    else:
        st.error("Model not loaded. Cannot make predictions.")

st.markdown("### About the Application")
st.info("This application uses a Random Forest Classifier model to recommend crops based on Nitrogen, Phosphorus, Potassium levels, Temperature, Humidity, pH, and Rainfall. The model was trained on the 'Crop_recommendation.csv' dataset.")

```
